```python
#convert positive to 1 and negative to 0
df['sentiment'] = df['sentiment'].replace(['positive', 'negative'], [1, 0])
```

```python
import openai
import pandas as pd
import json
import time


# Define a function to call the API and return the result
def get_sentiment(review):
    completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        max_tokens=1, #this is to prevent it from suddenly writing long outputs
        messages=[
            {
                "role": "user",
                "content": f"Do sentiment analysis on the following movie review, only respond with 1 (positive) or 0 (negative), nothing else: {review}"
            }
        ]
    )
    json_object = json.loads(str(completion))
    output = json_object['choices'][0]['message']['content']

    print(f"Review: {review}")
    print(f"Output: {output}")
    print("-"*50)  # print a divider for readability

    time.sleep(1)  # Pause for 1 second, sometimes we get some server errors if we don't do this
    return output

# Apply the function to the 'review' column and assign the result to the new column 'chatgpt'
df['chatgpt'] = df['review'].apply(get_sentiment)
```

```python
df
```

```python
import numpy as np
df['sentiment'] = df['sentiment'].astype(int)
df['chatgpt'] = df['chatgpt'].astype(int)
#df['check'] = np.where(df['sentiment'] == df['chatgpt'], 1, 0)

df.loc[df['sentiment'] == df['chatgpt'], 'check'] = 1
df.loc[df['sentiment'] != df['chatgpt'], 'check'] = 0

accuracy = df['check'].sum() / df.shape[0]
print('Accuracy:', accuracy)
```

```python
import pandas as pd
# Load spreadsheet
xl = pd.ExcelFile('/content/fertility_check_manual.xlsx')

# Load a sheet into a DataFrame by name
df = xl.parse('fert_solutions')
```

```python
df
```

```python
def create_manual_label(row):
    label_dict = {
        '友善職場': '1',
        '經濟支持': '2',
        '教保環境': '3',
        'other?': '4',
        'no specific policy direction offered': '5'
    }

    labels = [label_dict[col] for col in row.index if row[col] == 1]
    return ', '.join(labels)

df['manual_label'] = df.apply(create_manual_label, axis=1)
```

```python
# save
file_name = "manual_fertsolutions.csv"

# Write DataFrame to csv
df.to_csv(file_name, index=False)
```

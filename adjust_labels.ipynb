```python
# Load CSV file into a DataFrame
import pandas as pd
df = pd.read_csv('/content/fertsolutions_manual_label.csv')
```

```python
df = df.drop(columns='manual_label')
```

```python
df.columns
```

```python
def create_manual_label(row):
    label_dict = {
        'å‹å–„è·å ´': '1',
        'ç¶“æ¿Ÿæ”¯æŒ': '2',
        'æ•™ä¿ç’°å¢ƒ': '3',
        'other': '4',
        'no_policy': '5'
    }

    labels = []
    for col in label_dict:
        try:
            if float(row[col]) == 1.0:
                labels.append(label_dict[col])
        except ValueError:
            continue

    return ', '.join(labels)

df['manual_label'] = df.apply(create_manual_label, axis=1)
```

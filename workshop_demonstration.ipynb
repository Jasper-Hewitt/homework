```python
import pandas as pd
import random
from faker import Faker

# Initialize a faker generator
fake = Faker()

# Generate random names
names = [fake.name() for _ in range(10)]

# Generate random scores for each subject
math_scores = [random.randint(50, 100) for _ in range(10)]
lit_scores = [random.randint(50, 100) for _ in range(10)]
geo_scores = [random.randint(50, 100) for _ in range(10)]
hist_scores = [random.randint(50, 100) for _ in range(10)]

# Create a dataframe
df = pd.DataFrame({
    'student': names,
    'math': math_scores,
    'literature': lit_scores,
    'geography': geo_scores,
    'history': hist_scores
})

print(df)
```

```python
#I have a dataframe that looks like this (show dataframe)
#create a new column called 'average' that contains the average score for each student
df['average'] = df[['math', 'literature', 'geography', 'history']].mean(axis=1)
print(df)
```

```python
#plot the 'average' column in a sideways stackplot, descending, color: lightblue.
import matplotlib.pyplot as plt

# Sort the dataframe by 'average' column in descending order
df_sorted = df.sort_values('average', ascending=False)

# Create a horizontal bar plot
plt.figure(figsize=(10,8))
plt.barh(df_sorted['student'], df_sorted['average'], color='lightblue')

# Invert the y-axis to have the highest average at the top
plt.gca().invert_yaxis()

# Set labels
plt.xlabel('Average Score')
plt.ylabel('Student')
plt.title('Student Average Scores')

plt.show()
```

```python
# show a red line that contains the average of all student's average score
import matplotlib.pyplot as plt

# Sort the dataframe by 'average' column in descending order
df_sorted = df.sort_values('average', ascending=False)

# Calculate the overall average of student averages
overall_avg = df_sorted['average'].mean()

# Create a horizontal bar plot
plt.figure(figsize=(10,8))
plt.barh(df_sorted['student'], df_sorted['average'], color='lightblue')

# Add a vertical line indicating the overall average
plt.axvline(overall_avg, color='red', linestyle='--')

# Invert the y-axis to have the highest average at the top
plt.gca().invert_yaxis()

# Set labels
plt.xlabel('Average Score')
plt.ylabel('Student')
plt.title('Student Average Scores')

plt.show()
```
